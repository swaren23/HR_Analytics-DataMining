```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
matplotlib.style.use('ggplot')
myData=pd.read_csv('/home/yash/Downloads/HR_comma_sep.csv')
```

```python
myData.describe()
myData.sort_values(by='number_project',ascending=0).head(10)
myData.iloc[:,:]
myData
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(18,8))
g=sns.distplot(myData['satisfaction_level'],ax=axes[0],color='orange')
g=sns.distplot(myData['last_evaluation'],ax=axes[1])
#g=sns.distplot(myData['number_project'],ax=axes[2])
g=sns.distplot(myData['average_montly_hours'],ax=axes[2],color='purple')
#g=sns.distplot(myData['time_spend_company'],ax=axes[4])

plt.show()
```

```python
fig, axes = plt.subplots(ncols=4,figsize=(12,6))
g = sns.countplot(myData["left"], ax=axes[0])
g = sns.countplot(myData["Work_accident"], ax=axes[1])
g = sns.countplot(myData["number_project"], ax=axes[2])
g = sns.countplot(myData["time_spend_company"], ax=axes[3])



plt.show()
```

```python
fig, axes = plt.subplots(ncols=2,figsize=(12,6))
g = sns.countplot(myData["sales"], ax=axes[0])
plt.setp(g.get_xticklabels(), rotation=90)
g = sns.countplot(myData["salary"], ax=axes[1])
plt.show()
```

```python
g = sns.heatmap(myData.corr())
plt.title('correlation between different variables')
plt.show()
```

```python
data_stay=myData[myData['left']==0]
data_left=myData[myData['left']==1]

fig, axes = plt.subplots(ncols=3,figsize=(18,6))
sns.factorplot(y="satisfaction_level",x="left",data=myData,kind="box", ax=axes[0])
axes[1].hist(data_stay["satisfaction_level"],bins=50,label="Stay",alpha=0.7)
axes[1].hist(data_left["satisfaction_level"],bins=50,label="Left",alpha=0.7)
axes[1].set_xlabel("Satifaction level")
axes[1].set_ylabel("Count")
axes[1].legend()


g = sns.kdeplot(data=data_stay["satisfaction_level"],color='b',shade=True,ax=axes[2])
g = sns.kdeplot(data=data_left["satisfaction_level"],color='g',shade=True, ax=axes[2])
g.legend(["Stay","Left"])
g.set_xlabel('Satifsfaction level')
g.set_ylabel('Density')


plt.tight_layout()
plt.gcf().clear()
plt.show()
```

```python
fig, axes = plt.subplots(ncols=3,figsize=(18,6))
sns.factorplot(y="last_evaluation",x="left",data=myData,kind="box", ax=axes[0])
axes[1].hist(data_stay["last_evaluation"],bins=50,label="Stay",alpha=0.7)
axes[1].hist(data_left["last_evaluation"],bins=50,label="Left",alpha=0.7)
axes[1].set_xlabel("last_evaluation level")
axes[1].set_ylabel("Count")
axes[1].legend()


g = sns.kdeplot(data=data_stay["last_evaluation"],color='b',shade=True,ax=axes[2])
g = sns.kdeplot(data=data_left["last_evaluation"],color='g',shade=True, ax=axes[2])
g.legend(["Stay","Left"])
g.set_xlabel('last_evaluation')
g.set_ylabel('Density')


plt.tight_layout()
plt.gcf().clear()
plt.show()
```

```python
fig, axes = plt.subplots(ncols=1,figsize=(8,8))
#sns.factorplot(y="number_project",x="left",data=myData,kind="box", ax=axes[0])
axes.hist(data_stay["number_project"],label="Stay",bins=10,stacked=1)
axes.hist(data_left["number_project"],label="Left")
axes.set_xlabel("number_project")
axes.set_ylabel("Count")
axes.legend()

plt.show()
```
