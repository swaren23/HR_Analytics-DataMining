## Obtaining the Data

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('HR_comma_sep.csv', index_col=None)
```

## Data Pre-Processing

```python
# Check to see if there are any missing values in our data set
df.isnull().any()
```

```python
# Get a quick overview of what we are dealing with in our dataset
df.head()
```

```python
# Renaming certain columns for better readability
df = df.rename(columns={'satisfaction_level': 'satisfaction_level', 
                        'last_evaluation': 'latest_evaluation',
                        'number_project': 'projectCount',
                        'average_montly_hours': 'averageMonthlyHours',
                        'time_spend_company': 'yearsAtCompany',
                        'Work_accident': 'workAccident',
                        'promotion_last_5years': 'promotion_in_last_5_years',
                        'sales' : 'department',
                        'left' : 'left'
                        })
```

```python
# Move the reponse variable "left" to the front of the table
front = df['left']
df.drop(labels=['left'], axis=1,inplace = True)
df.insert(0, 'left', front)
df.head()
```

## Statistical Overview
***
The dataset has:
 - 14,999 employee observations and 10 features 
 - The company has a left rate of about 24%
 - Mean satisfaction of employees is 0.61
Features:
 - **satisfaction_level**: Satisfaction level of the employee on a scale of **0-1**.
 - **latest_evaluation**: The latest performance evaluation of the employee on a scale of **0-1**.
 - **projectCount**: Number of projects an employee is working on or has worked on.
 - **averageMonthlyHours**: Average number of hours spend by the employee over a month.
 - **yearsAtCompany**: Number of years spent in the company by the employee.
 - **workAccident**: If the employee has met with an accident at work(**0 for no, 1 for yes**)
 - **promotion_in_last_5_years**: If the employee has got a promotion in the last 5 years(**0 for no, 1 for yes**)
 - **salary**: Category of salary the employee falls into(**low, medium, high**)
 - **department**: The department to which the employee belongs
 - **left**: If the employee has left the company or not(**0 for no, 1 for yes**)

## Correlation Matrix and Heatmap
***
### Moderate Positively Correlated Features:
* projectCount vs latest_evaluation: 0.349333
* projectCount vs averageMonthlyHours: 0.417211
* latest_evaluation vs averageMonthly Hours: 0.339742

### Moderate Negatively Correlated Features:
* left vs satisfaction_level: -0.388375

### Summary 

* From the heatmap, there is a **positive(+)** correlation between projectCount, averageMonthlyHours, and latest_evaluation. Which could mean that the employees who spent more hours and did more projects were evaluated highly. 

* For the **negative(-)** relationships, 'left' and 'satisfaction' are highly correlated. It follows that people tend to leave a company more when they are less satisfied. 

```python
#Correlation Matrix
corr = df.corr()
_ , ax = plt.subplots( figsize =( 12 , 10 ) )
cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )
_ = sns.heatmap(
    corr, 
    cmap = cmap,
    square=True, 
    cbar_kws={ 'shrink' : .9 }, 
    ax=ax, 
    annot = True, robust=True,
    annot_kws = { 'fontsize' : 12 })
sns.plt.title('Heatmap of Correlation Matrix')
corr
```

##  Satisfaction_level vs Latest_evaluation
***

This is by far the most compelling graph

**Cluster 1:** Satisfaction was below **0.2** and evaluations were greater than **0.75**. Which could be a good indication that employees who left the company were good workers but felt horrible at their job.
